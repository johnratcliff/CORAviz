## CORA visualization

```python
import intake
import dask
import pandas as pd
import glob
import xarray as xr
import os
import natsort as natsorted
```

```python
catalog = intake.open_catalog('s3://noaa-nos-cora-pds/CORA_V1.1_intake.yml', storage_options={'anon':True})

ds = catalog['CORA-V1.1-fort.63-timeseries'].to_dask()
```

```python
dszeta = ds['zeta'].data
chunk1 = dszeta[0:2400,0:20000].compute()
```

```python
dszetanode0 = ds['zeta'].isel(node=0, time=slice(0,2400)).compute()
```

```python
for chunk_index, chunk in enumerate(dszeta.chunks):
    print(f"Chunk {chunk_index}:")
    print(chunk)  # This will show the dask array for that chunk
    # To see the actual values, you need to compute the chunk:
    print(chunk.compute())
```
